```python
import numpy as np
import pandas as pd
np.random.seed(41)
data={
    "x":np.arange(1,9),
    "y":np.random.randint(8,100,size=8)
}
df=pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.plot(df["x"],df["y"],marker='^',color= 'blue')
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot("random data set")
plt.grid(True)
plt.show()
```
