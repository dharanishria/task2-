```python
import numpy as np
import pandas as pd

np.random.seed(50)

data={
    "x" : np.arange(1,11),
    "y" : np.random.randint(14,100,size=10)
}
df=pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (5,5))
plt.plot(df["x"],df["y"])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
Products=['A','B','C','D']
Sales=[40,70,30,90]
plt.figure(figsize = (5,5))
plt.bar(Products,Sales,color='orange')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales analysis")
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np


x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 6]


plt.figure(figsize=(5,5))

plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title("Line Plot")
plt.subplot(2, 2, 2)
plt.scatter(x, y)
plt.title("Scatter Plot")

plt.subplot(2, 2, 3)
plt.bar(x, y)
plt.title("Bar Plot")

data = np.random.randn(50)
plt.subplot(2, 2, 4)
plt.hist(data)
plt.title("Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.randn(200)
plt.hist(data,edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Freq")
plt.show()
```
